```python
r = requests.post('https://api.dealroom.co/api/v1/companies/filters',
                auth = ('',''))
```

```python
r.status_code
```

```python
r_json = r.json()['items']
```

```python
r_json 
```

### CrÃ©ation d'un dataframe avec le titre des filtres

```python
filters_df = pd.DataFrame([d['key'] for d in r_json], columns = ['filters'])
```

```python
filters_df
```

```python
def get_categories(filt):
    categories = []
    for dic in r_json:
        if dic['key'] == filt:
            for d in dic['items']:
                categories.append(d['name'])
            return categories
    print('filter not found')
```

### Fonction get_categories qui retourne les catÃ©gories d'un filtre

Dictionnaire des filtres et leurs categories

```python
filters_dict = {}
for flt in filters_df.filters : 
    try : 
        filters_dict[flt] = get_categories(flt)
    except : 
        continue
```
